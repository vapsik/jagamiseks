Klassikaline DMD rakendus:

```python
X_gen = generation.copy()[:, :-1]
X_gen_prime = generation.copy()[:, 1:]
X = post_generation.copy()[:, :-1]
X_prime = post_generation.copy()[:, 1:]
X.shape
```

Tahame leida best-fit operaatori A, mis rahuldaks võrrandi X' = AX.

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)
S = np.diag(S)
```

SVD puhul ollakse huvitatud dimensionaalsuse (astaku) vähendamisest.

```python
# Plot singular values and cumulative sum
#plt.semilogy(np.diag(S))
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.grid()
plt.show()
plt.semilogy(np.diag(S))
plt.grid()
plt.show()

print(np.diag(S)[6])
```

```python
# olgu meie uuritav astak r =...
r = 500
U_tilde = U[:,0:r] 
S_tilde = S[0:r,0:r]
VT_tilde = VT[0:r,:]
```

```python
A_tilde = U_tilde.T @ X_prime @VT_tilde.T @ np.linalg.inv(S_tilde)

plt.imshow(A_tilde[:, :])
plt.show()
print(np.diag(A_tilde).max())

print(np.diag(A_tilde).min())

print(np.diag(A_tilde).mean())
```

```python
Lambda, W = np.linalg.eig(A_tilde)

#DMD moodid:
Phi = X_prime @ VT_tilde.T @ np.linalg.inv(S_tilde)

residual = X_prime - U_tilde @ A_tilde @ U_tilde.T @ X
frobenius_norm = np.linalg.norm(residual, 'fro')
print(f"X' ~= AX lähenduse Frobeniuse norm: {frobenius_norm}")

plt.imshow(Phi)
plt.show()
```

```python
plt.plot(y)
plt.title("Andmestiku ristlõige @ 500. samm")
plt.grid()
plt.show()
```

```python
c = 3e8
L = 1
f_min = c / L  # miinimum sagedus, mis vastab ühele perioodile ruumipiirkonnas
f_max = c / (0.1 * L)  # maksimum sagedus, mis vastab kümnele perioodile ruumipiirkonnas
f_0 = (f_min + f_max) / 2
N = 1000
dx = L/N
dt = dx/c
f_0, dx
```

```python
f_0/1e9
```

```python
# numpy FFT-ga sagedusspektri arvutamine
fft_values = np.fft.fft(y)

n = len(y)*dx
frequencies = np.fft.fftfreq(n)/c
magnitudes = np.abs(fft_values)

plt.figure(figsize=(24, 6))
plt.plot(np.abs(frequencies[:50), magnitudes[:50])

plt.title('Sageduste spekter')
plt.xlabel('Sagedus (Hz)')
plt.ylabel('Amplituud')
plt.grid()
plt.show()
```
